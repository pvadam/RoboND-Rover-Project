## Test matrix (bw image)

```python
# Import numpy
import numpy as np
# Import some packages from matplotlib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

img = np.random.rand(5, 5) 
img = img * 255
img = img.round(0)
print(img.shape)
print(type(img))
print(img.min())
print(img.max())
print(img.mean())
print(img.size)
print(img[0, 1])
img

plt.imshow(img, cmap='Greys',  interpolation='nearest')
plt.show()
```

## Masking

```python
masked = np.copy(img)
mask = masked < 205
masked[mask] = 0
masked

plt.imshow(masked, cmap='Greys',  interpolation='nearest')
plt.show()
```

## Fancy indexing

```python
indexed = np.copy(img)
inds_r = np.arange(len(indexed))
inds_c = 4 * inds_r % len(indexed)
indexed[inds_r, inds_c] = 0
print(inds_r)
print(inds_c)
indexed

plt.imshow(indexed, cmap='Greys',  interpolation='nearest')
plt.show()
```

## Circle mask

```python
circle = np.random.rand(512, 512) 
nrows, ncols = circle.shape
row, col = np.ogrid[:nrows, :ncols]
cnt_row, cnt_col = nrows / 2, ncols / 2
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > (nrows / 2)**2)
circle[outer_disk_mask] = 0

plt.imshow(circle, cmap='Greys',  interpolation='nearest')
plt.show()
```

## Half circle mask

```python
circle_complex = np.random.rand(512, 512) 
nrows_cpx, ncols_cpx = circle_complex.shape
row_cpx, col_cpx = np.ogrid[:nrows_cpx, :ncols_cpx]
cnt_row_cpx, cnt_col_cpx = nrows_cpx / 2, ncols_cpx / 2
outer_disk_mask_cpx = ((row_cpx - cnt_row_cpx)**2 + (col_cpx - cnt_col_cpx)**2 > (nrows_cpx / 2)**2)

lower_half = row_cpx > cnt_row_cpx
lower_half_disk = np.logical_and(lower_half, outer_disk_mask_cpx)
circle_complex[lower_half_disk] = 0

plt.imshow(circle_complex, cmap='Greys',  interpolation='nearest')
plt.show()
```
