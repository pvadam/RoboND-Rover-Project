```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2

image = mpimg.imread('sample_grid.jpg')

def perspect_transform(img, src, dst):

    # Get transform matrix using cv2.getPerspectivTransform()
    M = cv2.getPerspectiveTransform(src, dst)
    # Warp image using cv2.warpPerspective()
    # keep same size as input image
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    # Return the result
    return warped

# TODO:
# Define a box in source (original) and 
# destination (desired) coordinates
# Right now source and destination are just 
# set to equal the four corners
# of the image so no transform is taking place
# Try experimenting with different values!
source = np.float32([[53, 126], 
                 [124, 96], 
                 [199, 96], 
                 [270, 126]])
destination = np.float32([[155, 159], 
                 [155, 149], 
                 [165, 149], 
                 [165, 159]])

warped = perspect_transform(image, source, destination)
# Draw Source and destination points on images (in blue) before plotting
cv2.polylines(image, np.int32([source]), True, (0, 0, 255), 3)
cv2.polylines(warped, np.int32([destination]), True, (0, 0, 255), 3)
# Display the original image and binary               
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 6), sharey=True)
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=40)

ax2.imshow(warped, cmap='gray')
ax2.set_title('Result', fontsize=40)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
plt.show() # Uncomment if running on your local machine
```

## Color threshold & perspective transform

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Uncomment the next line for use in a Jupyter notebook
%matplotlib inline
import numpy as np
import cv2

# Read in the same sample image as before
image = mpimg.imread('sample_grid.jpg')

def perspect_transform(img):
    
    src = np.float32([[53, 126], 
                 [124, 96], 
                 [199, 96], 
                 [270, 126]])
    dst = np.float32([[155, 159], 
                     [155, 149], 
                     [165, 149], 
                     [165, 159]])

    # Get transform matrix using cv2.getPerspectivTransform()
    M = cv2.getPerspectiveTransform(src, dst)
    # Warp image using cv2.warpPerspective()
    # keep same size as input image
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    # Return the result
    return warped

def color_thresh(img, rgb_thresh=(0, 0, 0)):
    # Create an empty array the same size in x and y as the image 
        # but just a single channel
    binary_image = np.zeros((img.shape[0],img.shape[1]), dtype=int)
    
    # Apply the thresholds for RGB and 
        # assign 1's where threshold was exceeded
    mask = (img[:,:,0] > rgb_thresh[0]) \
        & (img[:,:,1] > rgb_thresh[1]) \
        & (img[:,:,2] > rgb_thresh[2])
    binary_image[mask] = 1
    return binary_image

# Assume you have already defined perspect_transform() and color_thresh()
warped = perspect_transform(image)
colorsel = color_thresh(warped, rgb_thresh=(160, 160, 160))

# Plot the result
plt.imshow(colorsel, cmap='gray')
plt.show()
```

```python
ypos, xpos = colorsel.nonzero()
print(xpos, ypos)
print(xpos[0], ypos[0])
print(colorsel.shape[1])

x_pixel = colorsel.shape[0] - ypos
y_pixel = colorsel.shape[1] / 2 - xpos
print(x_pixel, y_pixel)
print(x_pixel[0], y_pixel[0])

# plt.plot(xpos, ypos, '.')
plt.plot(x_pixel, y_pixel, '.')
plt.xlim(0, 160)
plt.ylim(-160, 160)
plt.show()
```

## Mapping to world coordinates

### Rotation

```python
import numpy as np
yaw = -30
# yaw angle is recorded in degrees so first convert to radians
yaw_rad = yaw * np.pi / 180
x_rotated = x_pixel * np.cos(yaw_rad) - y_pixel * np.sin(yaw_rad)
y_rotated = x_pixel * np.sin(yaw_rad) + y_pixel * np.cos(yaw_rad)

plt.plot(x_rotated, y_rotated, '.')
plt.xlim(0, 160)
plt.ylim(-160, 160)
plt.show()
```

### Transform

```python
import numpy as np
scale = 10
x_rovpos = 190
y_rovpos = 190

x_world = np.int_(x_rovpos + (x_rotated / scale))
y_world = np.int_(y_rovpos + (y_rotated / scale))

# ignore pixels outside the map
world_size = 200
x_pix_world = np.clip(x_world, 0, world_size - 1)
y_pix_world = np.clip(y_world, 0, world_size - 1)

plt.plot(x_pix_world, y_pix_world, '.')
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.show()
```
